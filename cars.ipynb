```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
data=pd.read_csv("cars-1.csv")
```

```python
data.head()
```

```python
x=data['model']
y=data['hp']
plt.figure(figsize=(25,25))
plt.plot(x,y)
plt.xlabel("Models of the Cars")
plt.ylabel("Horse Power of Cars")
plt.title("Model Names vs HorsePower")
plt.show()
```

```python
x=data["carb"]
y=data["gear"]
plt.bar(x,y,color='lime',width=0.5)
plt.xlabel("Number of carburetors")
plt.ylabel("Number of forward gears")
plt.title("carbs vs gear")
plt.show()
```

```python
x=data["wt"]
plt.hist(x,30,density=True,histtype='bar',color='green')
plt.xlabel("Weight of the cars")
plt.ylabel("Count")
plt.title("Histogram for the weight values")
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['model_factor'] = pd.factorize(data.model)[0]
data['cyl_factor'] = le.fit_transform(data.cyl)
data.head()
```

```python
x=data["cyl_factor"]
y=data["model_factor"]
plt.pie(y,labels=x)
plt.show()
```

```python
x=data["am"]
y=data["carb"]

plt.fill_between(x, y, color="skyblue", alpha=0.4)
plt.plot(x,y, color="Slateblue", alpha=0.6, linewidth=2)

plt.xlabel("Transmission")
plt.ylabel("Number of Carburetors")
plt.title("Transmission vs Number of carburetors")
```
